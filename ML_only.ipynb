# ML

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

```python
df=df3[selected_features]
```

```python
df.info()
```

```python
df.columns
```

```python
df['avg_salary'].unique()
```

```python
columns_to_encode = ['Job Title', 'Location', 'Size',
       'Type of ownership', 'Industry', 'Sector', 'Revenue']


for col in columns_to_encode:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    print('Label encoding complete.')
    filename = "encoder_pickles\LabelEncoder_" + col.replace(" ", "_")
    pickle.dump(le, open(filename, 'wb'))
```

```python
df.info()
```

```python
df.head()
```

```python
X = df.drop('avg_salary', axis=1)
y = df['avg_salary']
```

```python
print(X.shape, y.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=123)
print(X_train.shape, X_test.shape)
```

```python
print(X_test.columns)
X_test.head()
```

```python
y_test.head()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
import sklearn.metrics as metrics
```

**Linear Regression Model**

```python
lr = LinearRegression()
lr.fit(X_train, y_train) 
y_pred = lr.predict(X_test)
```

```python
# y_pred
```

```python
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

```python
filename = 'models\linear_reg_model.sav'
pickle.dump(lr, open(filename, 'wb'))
```

**SVR Model**

```python
svr = SVR()
svr.fit(X_train, y_train)
y_pred = svr.predict(X_test)
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

```python
filename2 = 'models\svr_model.sav'
pickle.dump(svr, open(filename2, 'wb'))
```

**KNN Model**

```python
knn = KNeighborsRegressor()
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

```python
# Storing the KNN Model
filename3 = 'models\knn_model.sav'
pickle.dump(knn, open(filename3, 'wb'))
```
